```python
import matplotlib.pyplot as plt

def showChart(dataFrame, xy, xyDesc, title, kind, xterms, bins): 
    chart = None
    if (kind == "bar"):
        chart = dataFrame.groupby(xy[0])[xy[1]].sum()
        chart = chart.plot(x=xy[0], y=xy[1], kind=kind)
    elif (kind == "hist"):
        chart = dataFrame[dataFrame[xy[0]] == xterms[0]][xy[1]].hist(alpha = 0.7, bins=bins)
        chart = dataFrame[dataFrame[xy[0]] == xterms[1]][xy[1]].hist(alpha = 0.7, bins=bins)
        plt.legend(list(xterms))
    elif (kind == "scatter" or kind == "line"):
        chart = dataFrame.plot(x=xy[0], y=xy[1], kind=kind)
    else:
        print("Chart must be one of the following kinds: ", ['line', 'bar', 'scatter', 'hist', 'barh', 'kde', 'density', 'area', 'box', 'pie', 'hexbin'])
        return None
    chart.set_xlabel(xyDesc[0])
    chart.set_ylabel(xyDesc[1])
    plt.title(title)
    plt.show()
    plt.close()


xy = ("type", "avg_price")
xyDesc = ("Average price", "Number of avocado sold")
title = "Average price vs Number of avocados sold"
kind = "hist"
xterm = ("conventional", "organic")
bins = 20

showChart(avocados, xy, xyDesc, title, kind, xterm, bins)
```
